```python
import pandas as pd

# List of CSV files for each scenario and corresponding scenario names
csv_files = [
    ('Results_1_test_table_summed.csv', '10s static P*'),
    ('Results_2_test_table_summed.csv', '2s static P*'),
    ('Results_3_test_table_summed.csv', '0.5s static P*'),
    ('Results_4_test_table_summed.csv', '10s dynamic P*'),
    ('Results_5_test_table_summed.csv', '2s dynamic P*'),
    ('Results_6_test_table_summed.csv', '0.5s dynamic P*'),
]

# Read the data from all CSV files and add a new column for the scenario name
all_data = []
for file, scenario_name in csv_files:
    scenario_data = pd.read_csv(file)
    scenario_data['Scenario'] = scenario_name
    all_data.append(scenario_data)

# Concatenate the data into a single DataFrame
all_data = pd.concat(all_data)

# Split the data by trader and write to new CSV files
for trader_name, trader_data in all_data.groupby('Trader'):
    trader_data.to_csv(f'{trader_name}_data.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the data from the CSV file
data = pd.read_csv("trader_data/GVWY_data.csv")

# Create the plot
plt.figure(figsize=(14, 8))
sns.lineplot(data=data, x='Market Share', y='Profit', hue='Scenario', marker="o")
plt.title('Profit vs. Market Share for Trader GVWY by Scenario')
plt.xlabel('Market Share (%)')
plt.ylabel('Profit')
plt.legend(title='Scenario', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
