```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV data from the file
file_name = "Results_1_test_table_summed.csv"
df = pd.read_csv(file_name)

# Create a line plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Market Share', y='Profit', hue='Trader')

# Add reference lines at specific y-values
y_values = [25, 50, 75, 100, 125, 150, 175, 200]  # You can modify these values as needed
for y in y_values:
    if y == 100:
        plt.axhline(y=y, color='gray', linestyle='--', linewidth=2)  # Bold line at y=100
    else:
        plt.axhline(y=y, color='gray', linestyle='--')

plt.title('Market Share vs. Profit for Each Trader with static equilibrium at a 10s batch interval ')
plt.xlabel('Market Share')
plt.ylabel('Profit')
plt.legend(title='Trader')
plt.show()
```
